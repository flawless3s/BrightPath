```python
from datasets import load_dataset
from sentence_transformers import SentenceTransformer
```

```python
# Load pre-trained SBERT model
text_encoder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
ds = load_dataset("SuperAGI/GUIDE", split="train[:5000]") # Loading 5000 rows
```

```python
previous_actions = ds["previousAction"] # Isolating previousAction column
```

```python
text_embeddings = text_encoder.encode(previous_actions, convert_to_numpy=True, show_progress_bar=True)

print(text_embeddings.shape)   # (num_actions, 384)
print(text_embeddings[0]) # sentence embedding of the first row of column previousAction
```
